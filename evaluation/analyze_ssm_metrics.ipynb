# Analysis of SSM Metrics

This notebook analyzes the metrics computed from the Self-Similarity Matrices (SSMs). We will aggregate the metrics for the xLSTM model and display the mean and standard deviation for each metric.

```python
import pandas as pd
import numpy as np
```

```python
# Load the metrics CSV
df = pd.read_csv('ssm_metrics.csv')
```

```python
# Add a 'Model' column. Since all files are from xLSTM, we can just assign 'xLSTM' to all rows.
df['Model'] = 'xLSTM'
```

```python
# Define the metrics to aggregate
metrics = ['avg_offdiag', 'rep_density', 'struct_entropy', 'block_coherence']

# Accumulate results
results_list = []
unique_models = df['Model'].unique()

for model in unique_models:
    model_df = df[df['Model'] == model]
    res = {'Model': model}
    for metric in metrics:
        m_val = model_df[metric].mean()
        s_val = model_df[metric].std()
        res[metric] = f"{m_val:.3f} ± {s_val:.3f}"
    results_list.append(res)

results = pd.DataFrame(results_list)

# Display the results
print("Aggregated Metrics (Mean ± Std Dev):")
print(results.to_markdown(index=False))
```

```python
# Save the aggregated results to a CSV file
results.to_csv('aggregated_ssm_metrics.csv', index=False)
print("\nAggregated results saved to 'aggregated_ssm_metrics.csv'")
```
