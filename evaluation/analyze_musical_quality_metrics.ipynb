# Analysis of Musical Quality Metrics

This notebook loads the musical quality metrics from `musical_quality_metrics_xlstm_84000.csv` and computes summary statistics (mean ± standard deviation).

```python
import pandas as pd
import numpy as np
```

```python
# Load the metrics CSV
csv_path = "musical_quality_metrics_xlstm_84000.csv"
df = pd.read_csv(csv_path)

print(f"Loaded {len(df)} rows from {csv_path}")
df.head()
```

```python
# Filter out rows with errors if any
if 'error' in df.columns:
    df_clean = df[df['error'].isna() | (df['error'] == "") | (df['error'].astype(str) == "nan")].copy()
    print(f"Rows after filtering errors: {len(df_clean)}")
else:
    df_clean = df.copy()
    
# Select numeric columns for aggregation
numeric_cols = [
    "bars_N",
    "key_stability", 
    "key_change_rate", 
    "chord_diversity", 
    "pitch_range", 
    "stepwise_ratio", 
    "note_density_mean", 
    "note_density_std", 
    "duration_entropy"
]

# Ensure they are numeric
for col in numeric_cols:
    if col in df_clean.columns:
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
```

```python
# Compute Mean and Std
summary_data = []

for col in numeric_cols:
    if col in df_clean.columns:
        mean_val = df_clean[col].mean()
        std_val = df_clean[col].std()
        
        summary_data.append({
            "Metric": col,
            "Mean": mean_val,
            "Std": std_val,
            "Formatted": f"{mean_val:.4f} ± {std_val:.4f}"
        })

summary_df = pd.DataFrame(summary_data)

# Display the summary table
pd.set_option('display.max_colwidth', None)
display(summary_df[['Metric', 'Formatted']])
```

```python
# Optional: Save the summary to a CSV/Markdown if needed
summary_df.to_csv("metrics_summary_xlstm_84000.csv", index=False)
print(summary_df[['Metric', 'Formatted']].to_markdown(index=False))
```
